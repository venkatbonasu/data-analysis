```python
import pandas as pd
from transformers import pipeline
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_excel("Final Cleaned_Nescafe_Reviews (1).xlsx")
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
sentiments = []
scores = []

for Review in df['Review']:
    result = sentiment_pipeline(Review)[0]
    sentiments.append(result['label'])
    scores.append(result['score'])


df['sentiment'] = sentiments
df['sentiment_score'] = scores

ratings = []

for sentiment, score in zip(df['sentiment'], df['sentiment_score']):
    if sentiment == 'POSITIVE':
        if score >= 0.95:
            ratings.append(5)
        elif score >= 0.90:
            ratings.append(4)
        else:
            ratings.append(3)
    elif sentiment == 'NEGATIVE':
        if score >= 0.85:
            ratings.append(2)
        else:
            ratings.append(1)
    else:
        ratings.append(3)

df['rating'] = ratings


print(df)
```

```python
df.to_excel('New Sentiment_Analyzed_Nescafe_Reviews.xlsx')
```
